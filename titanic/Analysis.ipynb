```python
def age_to_category(age):
    if age < 4:
        return 0 #"baby"
    elif age < 10:
        return 1 #"child"
    elif age < 21:
        return 2 #"teen"
    elif age < 33:
        return 3 #"young adult"
    elif age < 50:
        return 4 #"adult"
    return 5 #"elder"
```

```python
age_cats = {
  0: "baby",
  1: "child",
  2: "teen",
  3: "young adult",
  4: "adult",
  5: "elder"
}
```

```python
df['age_phase'] = df['age']
```

```python
# 1st way
for i in range(df.shape[0]):
    df['age_phase'].iloc[i] = age_to_category(df['age'].iloc[i])
```

```python
# 2nd way: Optimized Code
df['age_phase'] = df['age'].apply(age_to_category)
df['age_phase'].hist(bins=20, figsize=(5, 5))
```

```python
# for phase in df['age_phase'].unique():
#     print(age_cats[phase])

#     phase_df = df[ df['age_phase']==phase ]
#     survived = phase_df[phase_df['survived'] == 1]

#     survived_percentage = (survived.shape[0] / phase_df.shape[0]) * 100

#     print("Count: ", phase_df.shape[0])
#     print("Servived : ", "%.2f"% survived_percentage, '%')
#     print("\n====\n")
# ###### Optimized Code
# Precompute the survival counts and percentages to avoid redundant calculations
survival_counts = df.groupby('age_phase')['survived'].agg(['sum', 'count'])

# Iterate over the unique age phases and print the results
for phase in df['age_phase'].unique():
    phase_name = age_cats[phase]
    total_count = survival_counts.loc[phase, 'count']
    survived_count = survival_counts.loc[phase, 'sum']
    survived_percentage = (survived_count / total_count) * 100

    print(phase_name)
    print("Count: ", total_count)
    print("Survived: ", "%.2f" % survived_percentage, '%')
    print("\n====\n")
```
